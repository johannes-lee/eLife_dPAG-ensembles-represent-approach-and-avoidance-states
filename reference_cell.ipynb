```python
track, neur, behav = getdicts(foldnames, '230', 'epm')
calc0 = neur['C_raw']
vsort = np.flip(np.sort(calc0.var(axis=0)))
thresh = 0.1*vsort[2]

n = np.nonzero(vsort < thresh)[0][0]

plt.figure(figsize=(8, 5))
plt.scatter([0, 1, 2, 3, 4] + list(range(6, n)), list(vsort[0:5]) + list(vsort[6:n]), color='b', s=9)
plt.scatter([5], [vsort[5]], color='r', s=36)
plt.scatter(np.arange(n, vsort.size), vsort[n:vsort.size], color='m', s=16)
plt.legend(['kept', 'reference', 'excluded'])

plt.plot(0, thresh, 'k_', ms=2000)
plt.ylim(0, 140)
plt.yticks([0, thresh, 140], ['0', 'threshold', '140'])
plt.xticks([0, vsort.size-1], [1, vsort.size])
plt.xlabel('cell no.')
plt.ylabel('variance of raw dF/F (a.u.)')
plt.savefig('./elife_figs/reference_cell.svg')
```
